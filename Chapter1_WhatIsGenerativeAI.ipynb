# Chapter 1: What is Generative AI?

## Introduction
Generative AI (Gen AI) is a branch of artificial intelligence that focuses on creating new, synthetic data that mimics the properties of real-world data. These models generate text, images, audio, or even video by learning patterns and distributions from a dataset.

### Applications of Generative AI
- **Text Generation**: Content creation, translation, and summarization.
- **Image Creation**: AI-generated art, style transfer, and synthetic image generation.
- **Audio and Speech**: Music composition and voice synthesis.
- **Code Generation**: Writing code snippets or automating coding tasks.

---

## How Does Generative AI Work?
Generative models use various techniques to learn from training data and generate new samples. The most common types include:
1. **Generative Adversarial Networks (GANs)**: A generator and discriminator compete to create realistic data.
2. **Variational Autoencoders (VAEs)**: Learn latent representations of data for reconstruction and generation.
3. **Transformers**: Leverage attention mechanisms for tasks like text generation (e.g., GPT models).

## Example: Generating Synthetic Data

Generative AI can create new data points based on learned distributions. Let's generate synthetic data using Gaussian noise in Python.


```python
import numpy as np
import matplotlib.pyplot as plt

# Generate synthetic data using Gaussian distribution
def generate_synthetic_data(samples, mean, std_dev):
    return np.random.normal(mean, std_dev, samples)

# Parameters
samples = 1000
mean = 0
std_dev = 1

# Generate data
data = generate_synthetic_data(samples, mean, std_dev)

# Plot the data
plt.hist(data, bins=30, alpha=0.7, color='blue')
plt.title('Synthetic Data Generated with Gaussian Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```

## Exercise

### Task:
Write a Python function to generate synthetic data using a uniform distribution and visualize it as a histogram.

### Example:
Input:
- Number of samples: 500
- Range: [0, 10]

Output:
- A histogram of the generated data.

---

### Solution:
Below is a sample implementation for the exercise.


```python
# Generate synthetic data using a uniform distribution
def generate_uniform_data(samples, lower_bound, upper_bound):
    return np.random.uniform(lower_bound, upper_bound, samples)

# Parameters
samples = 500
lower_bound = 0
upper_bound = 10

# Generate data
uniform_data = generate_uniform_data(samples, lower_bound, upper_bound)

# Plot the data
plt.hist(uniform_data, bins=20, alpha=0.7, color='green')
plt.title('Synthetic Data Generated with Uniform Distribution')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()
```
